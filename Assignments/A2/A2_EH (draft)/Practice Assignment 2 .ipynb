```python
outcome  = 'age'
features = ['tot_inc','wgt']

X, y = survey[features], survey[outcome] 
```

```python
import statsmodels.api as sm

X_vars = X[['tot_inc','wgt']]
X_vars = sm.add_constant(X_vars, prepend=False)
X_vars.head()
y1 = survey['age']

m = sm.OLS(y1, X_vars)

res = m.fit()
print(res.summary()) 
```

```python
new_survey = survey[['wgt','tot_inc']]
```

```python
ols_df = pd.concat([y, X], axis=1)
ols_df.head()
```

```python
ols_df['pre_age'] = res.predict(X_vars)
ols_df.head()
```

```python
import statsmodels.api as sm

outcome ='gender'
y2 = survey[outcome]

logit_model=sm.Logit(y2,X_vars)
result = logit_model.fit()
print(result.summary2())
```

```python
ols_df['gender_pred'] = result.predict(X_vars)
ols_df.head()
```
